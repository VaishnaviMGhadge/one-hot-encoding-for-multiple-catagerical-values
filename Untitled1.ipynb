```python
#select columns used for one hot encoding
df=pd.read_csv('CarPrice_Assignment.csv',usecols=['CarName','fueltype','aspiration'
                                                 ,'doornumber','carbody','drivewheel','enginelocation',
                                                 'enginetype','cylindernumber','fuelsystem',
                                                 ])
df.head(10)
```

```python
#chech\k unique values
for col in df.columns:
    print(col,':',len(df[col].unique()),'labels')
```

```python
#check how many new columns are added after one hot encoding
pd.get_dummies(df,drop_first=True).shape
```

```python
#let us find the top most frequent catagaries for the variable CarName
df.CarName.value_counts().sort_values(ascending=False).head(10)
```

```python
#let us make a list with most frequent catagaries 
top_10=[x for x in df.CarName.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
#make 10 binary varibles
for label in top_10:
    df[label]=np.where(df['CarName']==label,1,0)
df[['CarName']+top_10].head(80)
```

```python
#our next target is to get dummy binaries for  all catagarical features
def one_hot_encode(df,variable,top_x_labels):
    
    for label in top_x_labels:
        df[variable+'_'+label]=np.where(df[variable]==label,1,0)
#read dataset again
df=pd.read_csv('CarPrice_Assignment.csv',usecols=['CarName','fueltype','aspiration'
                                                 ,'doornumber','carbody','drivewheel','enginelocation',
                                                 'enginetype','cylindernumber','fuelsystem',
                                                 ])

#encode CarName into 10 most frequent catagaries
one_hot_encode(df,'CarName',top_10)
df.head(10)
```

```python
#now create dummies for remaining aslo
top_2=[x for x in df.fueltype.value_counts().sort_values(ascending=False).head(2).index]
top_2
```

```python
one_hot_encode(df,'fueltype',top_2)
df.head(10)
```
